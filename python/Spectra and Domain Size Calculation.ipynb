```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import dedalus.public as de
from filter_field_iso import filter_field
```

```python
nx = 256
ny = 256
```

```python

x = de.Fourier('x',nx)
y = de.Fourier('y',ny)
```

```python
d = de.Domain([x,y],grid_dtype='float')
```

```python
phi = d.new_field()
```

```python
seed = 100
ampl = 1e-3
```

```python
# Create initial conditions
gshape = d.dist.grid_layout.global_shape(scales=d.dealias)
slices = d.dist.grid_layout.slices(scales=d.dealias)
rand = np.random.RandomState(seed=seed)
noise = rand.standard_normal(gshape)[slices]

phi['g'] = ampl*noise
filter_field(phi)
```

```python
plt.figure()
plt.imshow(np.roll(phi['g'],10,axis=1))
```
